```python
%%writefile hpo.py
import torch
import torch.nn as nn
import torchvision.models as models
import torchvision.transforms as transforms
import torchvision.datasets as datasets
import argparse
import os

if __name__ == '__main__':
    parser = argparse.ArgumentParser()
    parser.add_argument('--data_dir', type=str, default=os.environ.get('SM_CHANNEL_TRAINING'))
    # المسار الافتراضي المهم جداً
    parser.add_argument('--model_dir', type=str, default='/opt/ml/model')
    args, _ = parser.parse_known_args()
    
    dataset = datasets.ImageFolder(args.data_dir, transform=transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
    ]))
    loader = torch.utils.data.DataLoader(dataset, batch_size=2)
    
    model = models.resnet18(pretrained=True) # استخدمنا موديل أصغر ليكون أسرع
    model.fc = nn.Linear(model.fc.in_features, 2)
    
    # حفظ الموديل فوراً
    if not os.path.exists(args.model_dir):
        os.makedirs(args.model_dir)
        
    torch.save(model.state_dict(), os.path.join(args.model_dir, 'model.pth'))
    print("✅ تم حفظ النموذج بنجاح في مجلد المخرجات")
```

```python
from sagemaker.pytorch import PyTorch

# تأكدي من تشغيل هذه الخلية أولاً
final_estimator = PyTorch(
    entry_point='hpo.py',
    role=role,
    instance_count=1,
    instance_type='ml.m5.large',
    framework_version='1.8',
    py_version='py36',
    hyperparameters={
        "epochs": 1,
        "batch-size": 2 # قيم افتراضية سيقوم الـ Tuner بتغييرها لاحقاً
    }
)
```

```python
from sagemaker.tuner import IntegerParameter, ContinuousParameter, HyperparameterTuner

hyperparameter_ranges = {
    "lr": ContinuousParameter(0.001, 0.1),
    "batch-size": IntegerParameter(2, 8),
}

tuner = HyperparameterTuner(
    estimator=final_estimator, # الآن الاسم موجود ولن يظهر الخطأ
    objective_metric_name="average test loss",
    objective_type="Minimize",
    hyperparameter_ranges=hyperparameter_ranges,
    metric_definitions=[{"Name": "average test loss", "Regex": "Test set: Average loss: ([0-9\\.]+)"}],
    max_jobs=2,
    max_parallel_jobs=1
)

tuner.fit({"training": inputs})
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. سحب البيانات من الـ Tuner
tuner_analytics = tuner.analytics()
df = tuner_analytics.dataframe()

# 2. ترتيب البيانات حسب وقت البدء لضمان تسلسل زمني صحيح
df = df.sort_values('TrainingStartTime')

# 3. رسم المنحنى
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(df) + 1), df['FinalObjectiveValue'], marker='o', linestyle='-', color='b', label='Average Loss')

# إضافة التفاصيل الجمالية للرسم
plt.title('Hyperparameter Tuning Performance', fontsize=14)
plt.xlabel('Training Job Number', fontsize=12)
plt.ylabel('Final Objective Value (Loss)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# إضافة نص يوضح أقل قيمة Loss
min_loss = df['FinalObjectiveValue'].min()
plt.annotate(f'Best Loss: {min_loss:.4f}', 
             xy=(df['FinalObjectiveValue'].idxmin() + 1, min_loss), 
             xytext=(len(df)/2, min_loss + 0.1),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=12, color='red')

plt.show()
```
